### Returns by Year

### Calculating Equity Risk Premium

```python
diff_stocks_bills = returns['Stocks - Bills']
diff_stocks_bills.head()
```

```python
diff_stocks_bonds = returns['Stocks - Bonds']
diff_stocks_bonds.head()
```

```python
sum(diff_stocks_bills)
```

```python
sum(diff_stocks_bonds)
```

```python
stat.mean(diff_stocks_bills)
```

```python
stat.mean(diff_stocks_bonds)
```

```python
list=[["Stocks - Bills",round(sum(diff_stocks_bills),4),round(stat.mean(diff_stocks_bills),8),
         str(round((stat.mean(diff_stocks_bills)*100),2))+'%'],["Stocks - Bonds",
        round(sum(diff_stocks_bonds),4),round(stat.mean(diff_stocks_bonds),8),
        str(round((stat.mean(diff_stocks_bonds)*100),2))+'%']]
```

```python
list
```

```python
print('\033[1m'"                                 EQUITY RISK PREMIUM                                ")
print('\033[0m'"                     Sum of Difference | Average of Difference | Equity Risk Premium")
for item in list:
    print(item[0], " "*(10), item[1], " "*13, item[2], " "*12, item[3])
```
