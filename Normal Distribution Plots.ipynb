#### Part B: Histograms
##### Histogram for Normally Distributed Variable x
###### A) Showing Density

```python
N = 1000
x = nu.random.normal(size=N)
plot.hist(x, density=True, color='orange', alpha=0.5, edgecolor='blue', bins=33)
plot.title('Histogram for Normally Distributed Variable x', fontsize=18)
plot.xlabel('X Axis',fontsize=12)
plot.ylabel('Y Axis',fontsize=12)
plot.show()
```

```python
x1 = np.random.normal(0, 1, N)
x2 = np.random.normal(2, 1, N)
x3 = np.random.normal(1, 2, N)

density = dict(histtype='stepfilled', alpha=0.2, normed=True, bins=50, edgecolor='black')

plot.hist(x1, **density)
plot.hist(x2, **density)
plot.hist(x3, **density)

plot.title('Stepfilled Histogram for Normally Distributed Variable x', fontsize=20)
plot.xlabel('X Axis',fontsize=12)
plot.ylabel('Y Axis',fontsize=12)
plot.show()
```
